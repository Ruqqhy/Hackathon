```python
# encoding the Tv column
le = LabelEncoder()
df['TV'] = le.fit_transform(df['TV'])
df.head()
```

```python
# Encode the Influencer column
df['Influencer'] = le.fit_transform(df['Influencer'])
df.head()
```

```python
# create a bar chart for the distribution
sns.countplot(data=df)
plt.show()
```

```python
# create a pie chart for the distribution for influencer
plt.pie(df['Influencer'].value_counts(), labels=df['Influencer'].value_counts().index, autopct='%1.1f%%')
plt.title('Distribution of Influencer')
plt.show()
```
